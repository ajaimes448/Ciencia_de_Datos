```python
def bar_char_with_percentage(feature, df):
    survived = df[df['Survived'] == 1][feature].value_counts()
    dead = df[df['Survived'] == 0][feature].value_counts()
    total = survived + dead

    survived_percentage = (survived / total) * 100
    dead_percentage = (dead / total) * 100

    test = pd.DataFrame([survived_percentage, dead_percentage])
    test.index = ["Survived", "Dead"]

    ax = test.plot(kind="bar", stacked=True, figsize=(10, 5))

    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        ax.text(x + width/2,
                y + height/2,
                '{:.1f}%'.format(height),
                horizontalalignment='center',
                verticalalignment='center')

    plt.title(f'Sobrevivientes y Fallecidos por {feature} (Porcentaje)')
    plt.ylabel('Porcentaje')
    plt.xticks(rotation=0)
    plt.show()
```

## Creando un mapeo de los titulos
Reemplazamos los titulo enctrados por los soguientes valores
* MR : 0
* Miss:1
* Mrs: 2

```python
def replace_titles(data):
  title_mapping= {"Mr":0, "Miss":1, "Mrs":2}
  data["Title"]= data["Title"].map(title_mapping).fillna(3)
  return data
```

```python
total_embarked = df['Embarked'].value_counts()


fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
axes = axes.flatten()


for i, embarked in enumerate(total_embarked.index):
    subset = df[df['Embarked'] == embarked]
    counts = subset['Survived'].value_counts(normalize=True) * 100
    ax = sns.barplot(x=counts.index, y=counts.values, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Puerto: {embarked}')
    axes[i].set_xlabel('Sobrevivió')
    axes[i].set_ylabel('Porcentaje')
    axes[i].set_xticks([0, 1])
    axes[i].set_xticklabels(['No', 'Sí'])
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 1, f'{height:.1f}',
                ha='center', fontsize=10)

plt.suptitle('Porcentaje de Sobrevivientes por Puerto de Embarque', y=1.02)
plt.tight_layout()
plt.show()
```
