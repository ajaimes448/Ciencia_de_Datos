```python
ax=(estado.Population/1_000_000).plot.hist(figsize=(4,4))
ax.set_title("Population")
ax.set_xlabel("Population (millions)")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
ax=(estado["Murder.Rate"]).plot.hist(figsize=(4,4), color="lightblue")
ax.set_title("Tasa de Homicidios")
ax.set_xlabel("Murder.Rate")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
