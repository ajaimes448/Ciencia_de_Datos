```python
def plot_learning_curve(estimator,title,X,y,xlim=None,cv=None,n_jobs=1,train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Ejemplos de entrenamiento")
    plt.ylabel("Puntuación de validación")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Puntuación de entrenamiento")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Puntuación de validación")

    plt.legend(loc="best")
    return plt
```

```python
def bar_char_with_percentage(feature, df):
    survived = df[df['Survived'] == 1][feature].value_counts()
    dead = df[df['Survived'] == 0][feature].value_counts()
    total = survived + dead

    survived_percentage = (survived / total) * 100
    dead_percentage = (dead / total) * 100

    test = pd.DataFrame([survived_percentage, dead_percentage])
    test.index = ["Survived", "Dead"]

    ax = test.plot(kind="bar", stacked=True, figsize=(10, 5))

    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        ax.text(x + width/2,
                y + height/2,
                '{:.1f}%'.format(height),
                horizontalalignment='center',
                verticalalignment='center')

    plt.title(f'Sobrevivientes y Fallecidos por {feature} (Porcentaje)')
    plt.ylabel('Porcentaje')
    plt.xticks(rotation=0)
    plt.show()
```

```python
def bar_char_with_percentage_by_pclass_gender(df):

    survival_counts = df.groupby(['Pclass', 'Sex', 'Survived']).size().unstack(fill_value=0)
    total_counts = survival_counts.sum(axis=1)


    survival_percentage = (survival_counts[1] / total_counts) * 100
    death_percentage = (survival_counts[0] / total_counts) * 100

    plot_df = pd.DataFrame({'Sobrevivió (%)': survival_percentage, 'Falleció (%)': death_percentage})


    ax = plot_df.plot(kind='bar', stacked=True, figsize=(10, 6))

    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        ax.text(x + width/2,
                y + height/2,
                '{:.1f}%'.format(height),
                horizontalalignment='center',
                verticalalignment='center')

    ax.set_title('Sobrevivientes y Fallecidos por Clase de Billete y Género (Porcentaje)')
    ax.set_xlabel('(Clase de Billete, Género)')
    ax.set_ylabel('Porcentaje')
    ax.tick_params(axis='x', rotation=45)
    ax.legend(title='Estado')
    plt.tight_layout()
    plt.show()
```

```python
def alone_vs_accompanied_graph(df):


    df['IsAlone'] = (df['SibSp'] == 0) & (df['Parch'] == 0)


    alone_counts = df['IsAlone'].value_counts()


    total_passengers = len(df)
    alone_percentage = (alone_counts[True] / total_passengers) * 100
    accompanied_percentage = (alone_counts[False] / total_passengers) * 100


    plt.figure(figsize=(8, 6))
    sns.barplot(x=['Accompanied', 'Alone'], y=[accompanied_percentage, alone_percentage], palette='muted')


    plt.text(0, accompanied_percentage + 1, f'{accompanied_percentage:.1f}%', ha='center', color='black')
    plt.text(1, alone_percentage + 1, f'{alone_percentage:.1f}%', ha='center', color='black')


    plt.title('Pasajeros Solos vs. Acompañados')
    plt.xlabel('Estado')
    plt.ylabel('Porcentaje de Pasajeros')
    plt.xticks([0, 1], ['Acompañados', 'Solos'])
    plt.tight_layout()
    plt.show()
```
